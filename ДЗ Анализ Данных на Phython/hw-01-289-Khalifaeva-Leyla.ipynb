## Решение

__а) [5 баллов]__ Постройте гистограмму количества автобусов, отошедших от автостанции за сутки. Похоже ли визуально распределение на нормальное?

```python
n_obs = 10**4
np.random.seed(2023)

bus_time = np.random.choice([5, 10], size=(n_obs,int(720/5)), p=[0.5, 0.5])
bus_time_cumsum = np.cumsum(bus_time, axis=1)
number_of_buses = []
for i in bus_time_cumsum:
    i = i[i<=720]
    number_of_buses.append(len(i))
number_of_buses = np.array(number_of_buses)
number_of_buses

plt.figure(figsize=(25,15))
plt.hist(number_of_buses, bins=50)
plt.xlabel('Число автобусов',fontsize=24)
plt.ylabel('Количество дней',fontsize=24)
plt.title('Распределение числа автобусов, отошедших от автостанции за сутки',fontsize=30)
```

### Задача 2 (Парадокс Хуана Мануэля Родригеса Паррондо)

У Атоса, Портоса и Арамиса по 1000 франков. Атос постоянно ходит в казино А, где каждый раз выигрывает один франк с вероятностью $0.49$ и проигрывает один франк с вероятностью $0.51$. Портос ходит в казино Б, где ситуация интереснее :) 

- Если богатство посетителя кратно трём, то посетитель выигрывает франк с вероятностью $0.09$ и проигрывает один франк с вероятностью $0.91$
- Если богатство посетителя не кратно трём, то посетитель выигрывает франк с вероятностью $0.74$ и проигрывает один франк с вероятностью $0.26$

Арамис каждый раз выбирает казино А или казино Б равновероятно. Проведите $10^4$ симуляций эволюции благосостояния Атоса, Портоса и Арамиса.Каждая симуляция предусматривает $1000$ посещений казино. Все три игрока все деньги носят с собой и ни на что не тратят :)

__а) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Атоса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

## Решение


__а) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Атоса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

```python
n_obs = 10**4
np.random.seed(2023)

begin_wealth = 1000 * np.ones(n_obs)
win_loss_A = np.random.choice([1, -1], size=(1000, n_obs), p=[0.49, 0.51])
wealth_A = 1000 + np.cumsum(win_loss_A, axis = 0)
wealth_A = np.insert(wealth_A, 0, begin_wealth, axis = 0 )
wealth_mean_A= np.sum(wealth_A, axis=1) / n_obs


plt.figure(figsize=(20,10))
plt.plot(wealth_A[:, :10], linewidth=0.7)
plt.xlabel('Номер посещения казино',fontsize=24)
plt.ylabel('Богатство Атоса',fontsize=24)
plt.title('Траектории изменения богатства Атоса',fontsize=25)
plt.plot(wealth_mean_A, color='r', linestyle='dashed', linewidth=6, label = 'усредненная')
plt.legend(fontsize=24)
```

## Вывод:
_Выбранная Атосом стратегия похода в казино А в среднем приводит к постепенному уменьшению его богатства на 20 франков, несмотря на возможные в некоторых случаях выигрыши._


__б) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Портоса, по горизонтали —  номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

```python
from tqdm import tqdm

n_obs = 10**4
np.random.seed(2023)
wealth_B = 1000 * np.ones(n_obs)

wealth_general = [list(wealth_B)]
for visit in tqdm(range(1,1001)):
    wealth_after = []
    for i in range(len(wealth_B)):
        if wealth_B[i] % 3 == 0:
            win_loss_B = np.random.choice([1, -1], size=1, p=[0.09, 0.91])
        else:
            win_loss_B = np.random.choice([1, -1], size=1, p=[0.74, 0.26])
        wealth_B[i] = wealth_B[i] + int(win_loss_B)
        wealth_after.append(wealth_B[i])
    wealth_general.append(wealth_after)
result = np.array(wealth_general)
result
```

```python
result_mean= np.sum(result, axis=1) / n_obs

plt.figure(figsize=(20,10))
plt.plot(result[:, :10], linewidth=0.7)
plt.xlabel('Номер посещения казино',fontsize=24)
plt.ylabel('Богатство Портоса',fontsize=24)
plt.title('Траектории изменения богатства Портоса',fontsize=25)
plt.plot(result_mean, color='r', linestyle='dashed', linewidth=6, label = 'усредненная')
plt.legend(fontsize=24)
```

## Вывод:
_Выбранная Портосом стратегия похода в казино Б так же в среднем приводит к постепенному уменьшению его богатства, но на меньшую сумму, около 18 франков, несмотря на возможные в некоторых случаях выигрыши._

__в) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Арамиса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.
    

```python
from tqdm import tqdm

n_obs = 10**4
np.random.seed(2023)

wealth = 1000 * np.ones(n_obs)
casino = np.random.choice(['A', 'B'], size=1000, p=[0.5, 0.5])
for t in tqdm(range(len(casino))):
    if casino[t] == 'A':
        win_loss_A = wealth[-1] + np.random.choice([1, -1], size=n_obs, p=[0.49, 0.51])
        wealth = np.insert([win_loss_A], 0, wealth, axis = 0 )
    else:
        wealth_after = []
        for i in range(len(wealth[-1])):
            if wealth[-1][i] % 3 == 0:
                win_loss_B = np.random.choice([1, -1], size=1, p=[0.09, 0.91])
            else:
                win_loss_B = np.random.choice([1, -1], size=1, p=[0.74, 0.26])
            wealth_ =  wealth[-1][i] + int(win_loss_B)    
            wealth_after.append(wealth_)
        wealth = np.insert([wealth_after], 0, wealth, axis = 0 )
wealth = np.array(wealth)
wealth
```

```python
wealth_mean = np.sum(wealth, axis=1) / n_obs

plt.figure(figsize=(20,10))
plt.plot(wealth[:, :10], linewidth=0.7)
plt.xlabel('Номер посещения казино',fontsize=24)
plt.ylabel('Богатство Арамиса',fontsize=24)
plt.title('Траектории изменения богатства Арамиса',fontsize=25)
plt.plot(wealth_mean, color='r', linestyle='dashed', linewidth=6, label = 'усредненная')
plt.legend(fontsize=24)
```

## Вывод:
_Стратагия Арасмиса, основанная на выборе между двумя казино, более успешна и позволяет ему в среднем выигрывать около 7 франков, несмотря на возможные в некоторых случаях проигрыши._

### Задача 3 (Парадокс Берксона)

Предположим, что результаты ЕГЭ школьников по русскому и математике независимы и хорошо аппроксимируются 
нормальным распределением с ожиданием 60 и стандартным отклонением 10. 

УШЭ (Урюпинская Школа Экономики) ныне престижна и забирает себе всех школьников набравших более $n$ баллов в сумме по русскому и математике. 

Для каждого $n$ от $80$ до $160$ с шагом $5$ случайно создайте $10^4$ школьников и поделите их на прошедших и не прошедших в УШЭ.


__а) [5 баллов]__ Постройте график числа прошедших в УШЭ в зависимости от $n$. 

__б) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди прошедших в УШЭ.

__в) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди не прошедших в УШЭ.

## Решение

__а) [5 баллов]__ Постройте график числа прошедших в УШЭ в зависимости от $n$. 

```python
n_obs = 10**4
np.random.seed(2023)

number_of_passed = []
number_of_failed = []
for n in np.arange(80, 161, 5):
    math_res = sts.norm(loc=60, scale=10).rvs(n_obs)
    russ_res = sts.norm(loc=60, scale=10).rvs(n_obs)
    exam_results = math_res + russ_res
    passed = exam_results[exam_results >= n]
    failed = exam_results[exam_results < n]
    number_of_passed.append(len(passed))
    number_of_failed.append(len(failed))
np.array(number_of_passed), np.array(number_of_failed)
```

```python
plt.figure(figsize=(20,10))
plt.plot(np.arange(80, 161, 5), np.array(number_of_passed))
plt.xlabel('Проходной балл',fontsize=24)
plt.ylabel('Число студентов',fontsize=24)
plt.title('Число студентов, прошедших в УШЭ',fontsize=25)
```

## Вывод:
_С ростом проходных баллов число студентов, прошедших в УШЭ, сокращается._


__б) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди прошедших в УШЭ.

```python
n_obs = 10**4
np.random.seed(2023)

math_passed = []
russ_passed = []
correlation = []
for n in np.arange(80, 161, 5):
    math_res = sts.norm(loc=60, scale=10).rvs(n_obs)
    russ_res = sts.norm(loc=60, scale=10).rvs(n_obs)
    exam_results = math_res + russ_res
    correlation.append(np.corrcoef(math_res[exam_results >= n], russ_res[exam_results >= n])[0,1])

plt.figure(figsize=(20,10))
plt.plot(np.arange(80, 161, 5), np.array(correlation))
plt.xlabel('Проходной балл',fontsize=24)
plt.ylabel('Коэффициент корреляции',fontsize=24)
plt.title('Выборочная корреляция между результатами по русскому и математике среди прошедших в УШЭ',fontsize=25)
```

## Вывод:
_Выборочная корреляция между результатами по русскому и математике среди прошедших в УШЭ студентов отрицательная. При этом, с ростом проходного балла отрицательная корреляция растет(по модулю)._

```python
n_obs = 10**4
np.random.seed(2023)

math_passed = []
russ_passed = []
correlation = []
for n in np.arange(80, 161, 5):
    math_res = sts.norm(loc=60, scale=10).rvs(n_obs)
    russ_res = sts.norm(loc=60, scale=10).rvs(n_obs)
    exam_results = math_res + russ_res
    correlation.append(np.corrcoef(math_res[exam_results < n], russ_res[exam_results < n])[0,1])

plt.figure(figsize=(20,10))
plt.plot(np.arange(80, 161, 5), np.array(correlation))
plt.xlabel('Проходной балл',fontsize=24)
plt.ylabel('Коэффициент корреляции',fontsize=24)
plt.title('Выборочная корреляция между результатами по русскому и математике среди не прошедших в УШЭ',fontsize=25)
```
